```python
import pandas as pd
```

```python
citibike = pd.read_csv('citibike_18_19.csv',parse_dates=['date']).rename({'date':'Date'},axis=1).set_index(['Date','HOUR_BIN'],drop=True).drop('Unnamed: 0',axis=1)
green_taxi = pd.read_csv('data_green_taxi.csv',parse_dates=['Date']).set_index(['Date','HOUR_BIN'],drop=True)
yellow_taxi = pd.read_csv('data_yellow.csv',parse_dates=['Date']).set_index(['Date','HOUR_BIN'],drop=True).drop('Unnamed: 0',axis=1)
MTA = pd.read_csv('DF_MTA',parse_dates=['DATE']).rename({'DATE':'Date'},axis=1).set_index(['Date','HOUR_BIN'],drop=True).drop('Unnamed: 0',axis=1)
weather = pd.read_csv('weather_data.csv',parse_dates=['Date']).set_index(['Date','HOUR_BIN'],drop=True).drop('Unnamed: 0',axis=1)
```

```python
green_taxi = green_taxi.rename({'passenger_count':'passenger_countg'},axis=1)
```

```python
MTA
```

```python
Merged_Data = pd.concat([citibike,green_taxi,yellow_taxi,MTA,weather], axis=1)
```

```python
Merged_Data.head()
```

```python
Merged_Data.columns
```

```python
Merged_Data = Merged_Data.rename({'ENTRY':'MTAEntry_Y','EXIT':'MTAExit_Y','trip_count':'Citibike_Y','passenger_countg':'GreenTaxi_Y','passenger_count':'YellowTaxi_Y'},axis=1)
```

```python
Merged_Data = Merged_Data.reset_index(drop=False).sort_values(['Date','HOUR_BIN']).set_index(['Date','HOUR_BIN'],drop=True)
```
